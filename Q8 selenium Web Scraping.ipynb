```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(executable_path="D:\\chromedriver_win32\\chromedriver.exe")
```

```python
driver.get("https://flipkart.com")
```

```python
search_Item = driver.find_element_by_xpath("//input[@class='_3704LK']")
search_Item.send_keys("sneakers")
search_btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()
```

```python
Brands=[]
Product_Desc=[]
Price=[]
Discounts=[]
```

```python
url="https://www.flipkart.com/search?q=sneakers&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
driver.get(url)
```

```python
import time
```

```python
for i in range(0,10):
    for i in driver.find_elements_by_xpath("//div[@class='_2WkVRV']"):
        Brands.append(i.text)
    for j in driver.find_elements_by_xpath("//a[@class='IRpwTa']"):
        Product_Desc.append(j.text)
    for k in driver.find_elements_by_xpath("//div[@class='_30jeq3']"):
        Price.append(k.text)
    for l in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']"):
        Discounts.append(l.text)
time.sleep(5)
```
